#Caricamento del dataset scaricato da Kaggle con i vincitori e i nominati gli Oscar dal 1927 al 2020

```python
url = 'https://drive.google.com/file/d/1D6ML3iCG624j29_eTylUbUel_EeoCmvU/view?usp=sharing'
path = 'https://drive.google.com/uc?export=download&id=' + url.split('/')[-2]

Oscar_df = pd.read_csv(path)
```

#Rimozione di attributi e record non rilevanti alla fine del progetto

```python
#estraggo dal dataset caricato solo i film prodotti nel 2019
Oscar_df = Oscar_df[Oscar_df.year_film == 2019]
Oscar_df=Oscar_df.reset_index(drop=True)

#seleziono gli attributi rilevanti allo scopo del progetto 
Oscar_df=Oscar_df[['year_film', 'category', 'name', 'film', 'winner']]

#rimuovo i premi con valore del film nullo (premi onorari assegnati ad personalità)
Oscar_df.dropna(subset=['film'], inplace=True)
Oscar_df
```

#Scraping dal sito [www.oscars.org](www.oscars.org/oscars/ceremonies/2021) dei vincitori e nominati agli Oscar del 2021

```python
from bs4 import BeautifulSoup
import requests
base_url = "https://www.oscars.org/oscars/ceremonies/2021"
response = requests.get(base_url) 
html_page = response.text
html_tree = BeautifulSoup(html_page)
```

```python
elenco = html_tree.find_all('div', class_="view-grouping")
for elemento in elenco: 
  premi = elemento.find_all('h2')
  for premio in premi: 
    if "Actor" in premio.text or "Actress" in premio.text:
      attore = elemento.find('div', class_="views-row views-row-1 views-row-odd views-row-first views-row-last")
      Oscar_df=Oscar_df.append({'year_film': 2020,	'category' : premio.text,	'name':	attore.find('h4').text, 'film' : attore.find('span', class_="field-content").text,	'winner': True}, ignore_index=True)
      attore = elemento.find('div', class_="views-row views-row-1 views-row-odd views-row-first")
      Oscar_df=Oscar_df.append({'year_film': 2020,	'category' : premio.text,	'name':	attore.find('h4').text, 'film' : attore.find('span', class_="field-content").text,	'winner': False}, ignore_index=True)
      attore = elemento.find('div', class_="views-row views-row-2 views-row-even")
      Oscar_df=Oscar_df.append({'year_film': 2020, 'category' : premio.text,	'name':	attore.find('h4').text, 'film' : attore.find('span', class_="field-content").text,	'winner': False}, ignore_index=True)
      attore = elemento.find('div', class_="views-row views-row-3 views-row-odd")
      Oscar_df=Oscar_df.append({'year_film': 2020,	'category' : premio.text,	'name':	attore.find('h4').text, 'film' : attore.find('span', class_="field-content").text,	'winner': False}, ignore_index=True)
      attore = elemento.find('div', class_="views-row views-row-4 views-row-even views-row-last")
      if attore!=None:
        Oscar_df=Oscar_df.append({'year_film': 2020,	'category' : premio.text,	'name':	attore.find('h4').text, 'film' : attore.find('span', class_="field-content").text,	'winner': False}, ignore_index=True)
    elif "Original Song" in premio.text: 
      attore = elemento.find('div', class_="views-row views-row-1 views-row-odd views-row-first views-row-last")
      Oscar_df=Oscar_df.append({'year_film': 2020,	'category' : premio.text,	'name':	attore.find('h4').text, 'film' : attore.find('span', class_="field-content").text.strip("from").split(';')[0],	'winner': True}, ignore_index=True)
      attore = elemento.find('div', class_="views-row views-row-1 views-row-odd views-row-first")
      Oscar_df=Oscar_df.append({'year_film': 2020,	'category' : premio.text,	'name':	attore.find('h4').text, 'film' : attore.find('span', class_="field-content").text.strip("from").split(';')[0],	'winner': False}, ignore_index=True)
      attore = elemento.find('div', class_="views-row views-row-2 views-row-even")
      Oscar_df=Oscar_df.append({'year_film': 2020,	'category' : premio.text,	'name':	attore.find('h4').text, 'film' : attore.find('span', class_="field-content").text.strip("from").split(';')[0],	'winner': False}, ignore_index=True)
      attore = elemento.find('div', class_="views-row views-row-3 views-row-odd")
      Oscar_df=Oscar_df.append({'year_film': 2020,	'category' : premio.text,	'name':	attore.find('h4').text, 'film' : attore.find('span', class_="field-content").text.strip("from").split(';')[0],	'winner': False}, ignore_index=True)
      attore = elemento.find('div', class_="views-row views-row-4 views-row-even views-row-last")
      Oscar_df=Oscar_df.append({'year_film': 2020,	'category' : premio.text,	'name':	attore.find('h4').text, 'film' : attore.find('span', class_="field-content").text.strip("from").split(';')[0],	'winner': True}, ignore_index=True)                              
    else: 
      attore = elemento.find('div', class_="views-row views-row-1 views-row-odd views-row-first views-row-last")
      Oscar_df=Oscar_df.append({'year_film': 2020,	'category' : premio.text,	'name':	attore.find('span', class_="field-content").text, 'film' : attore.find('h4').text,	'winner': True}, ignore_index=True)
      attore = elemento.find('div', class_="views-row views-row-1 views-row-odd views-row-first")
      Oscar_df=Oscar_df.append({'year_film': 2020,	'category' : premio.text,	'name':	attore.find('span', class_="field-content").text, 'film' : attore.find('h4').text,	'winner': False}, ignore_index=True)
      attore = elemento.find('div', class_="views-row views-row-2 views-row-even")
      Oscar_df=Oscar_df.append({'year_film': 2020,	'category' : premio.text,	'name':	attore.find('span', class_="field-content").text, 'film' : attore.find('h4').text,	'winner': False}, ignore_index=True)
      attore = elemento.find('div', class_="views-row views-row-3 views-row-odd")
      Oscar_df=Oscar_df.append({'year_film': 2020,	'category' : premio.text,	'name':	attore.find('span', class_="field-content").text, 'film' : attore.find('h4').text,	'winner': False}, ignore_index=True)
      attore = elemento.find('div', class_="views-row views-row-4 views-row-even views-row-last")
      if attore!=None:
        Oscar_df=Oscar_df.append({'year_film': 2020,	'category' : premio.text,	'name':	attore.find('span', class_="field-content").text, 'film' : attore.find('h4').text,	'winner': False}, ignore_index=True)
      attore = elemento.find('div', class_="views-row views-row-6 views-row-even views-row-last")
      if attore!=None:
        Oscar_df=Oscar_df.append({'year_film': 2020,	'category' : premio.text,	'name':	attore.find('span', class_="field-content").text, 'film' : attore.find('h4').text,	'winner': False}, ignore_index=True)
      attore = elemento.find('div', class_="views-row views-row-7 views-row-odd views-row-last")
      if attore!=None:
        Oscar_df=Oscar_df.append({'year_film': 2020,	'category' : premio.text,	'name':	attore.find('span', class_="field-content").text, 'film' : attore.find('h4').text,	'winner': False}, ignore_index=True)
```

#Controllo della consistenza tra il dataset scaricato e quello ottenuto tramite scraping

```python
#Per aumentare la consistenza del dataset viene uniformato il modo in cui sono indicati i nomi dei premi
Oscar_df['category'] = Oscar_df['category'].map(lambda x: x.upper())

#Per agevolare l'integrazione con ulteriori dataset vengono rimossi eventuali spazi bianchi in fondo alle stringhe 
#ottenute tramite scraping 

Oscar_df['film'] = Oscar_df['film'].map(lambda x: x.lstrip())
Oscar_df['film'] = Oscar_df['film'].map(lambda x: x.strip('\n'))
Oscar_df['name'] = Oscar_df['name'].map(lambda x: x.lstrip())
Oscar_df['name'] = Oscar_df['name'].map(lambda x: x.strip('\n')) 
```

```python
Oscar_df.to_csv('Oscar2019-20.csv', index_label='Index')
#Abbiamo nominato l'indice esterno come index
```
