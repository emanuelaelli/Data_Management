```python
# TUTORIAL https://medium.com/@avivamazurek/basics-of-web-scrapping-how-to-scrape-data-off-of-box-office-mojo-5ec6c22dcca6
# sito da cui prelevare i dati https://www.boxofficemojo.com/year/world/2019/

import requests
from bs4 import BeautifulSoup as BS
import pandas as pd
```

## Incassi 2019

```python
url = "https://www.boxofficemojo.com/year/world/2019/"
page = requests.get(url)
soup = BS(page.content, 'html.parser')
titles = soup.find_all('td', class_="a-text-left mojo-field-type-release_group")
titles[0]
```

```python
titles[0].select('a')[0].string
```

```python
All_titles = []
for thing in titles:
  All_titles.append(thing.select('a')[0].string)
```

```python
money = soup.find_all('td', class_="a-text-right mojo-field-type-money")
money[0].string
```

```python
All_money = []
for n in range(0, len(money)):
  All_money.append(money[n].string)
# All_money
```

```python
Worldwide = []
Domestic = []
Foreign = []

for n in range(0, len(money), 3):
  Worldwide.append(All_money[n])

for m in range(1, len(money), 3):
  Domestic.append(All_money[m])

for p in range(2, len(money), 3):
  Foreign.append(All_money[p])
```

```python
incassi2019 = pd.DataFrame(list(zip(All_titles, Worldwide, Domestic, Foreign))) 
incassi2019.columns =['Titles', 'Worldwide', 'Domestic', 'Foreign']
incassi2019.to_csv('incassi2019.csv')

df_incassi2019 = pd.read_csv('incassi2019.csv')
df_incassi2019.drop('Unnamed: 0',axis=1, inplace=True)
# df_incassi2019
```

```python
def list_index(x):
  id = []
  for item in list(x):
      movie = ia.search_movie(item) # cerca il film
      if movie:
        id.append(movie[0].getID()) # trova l'id
      else:
        id.append(None)
  return id
    
id_movies = list_index(incassi['Titles'])
id_movies
```

```python
new_list=[]
old_list = id_movies

prefix_ADD = 'tt'

for x in old_list:
    if x != None:
      new_list.append(prefix_ADD + x)
    else:
      new_list.append(' ')

print(new_list)
```

```python
# aggiungo nuovo attributo titleId
incassi['titleId']=new_list
incassi
```
